Percentage by Flight Status

```python
import pandas as pd
from tabulate import tabulate

# Load the dataset
data = pd.read_csv('Airline.csv')

# Count the occurrences of each flight status
status_counts = data['Flight Status'].value_counts()

# Calculate the percentage of each flight status
total_flights = len(data)
status_percentage = (status_counts / total_flights) * 100

# Create a DataFrame for display
status_summary = pd.DataFrame({
    "Flight Status": status_counts.index,
    "Count": status_counts.values,
    "Percent": status_percentage.values
})

# Display the DataFrame in a markdown format using tabulate
print(tabulate(status_summary, headers='keys', tablefmt='pipe', showindex=False))
```

Visual Representation of Flight Status

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Airline.csv')

# Count the occurrences of each flight status
status_counts = data['Flight Status'].value_counts()

# Calculate the percentage of each flight status
total_flights = len(data)
status_percentage = (status_counts / total_flights) * 100

# Create a DataFrame for display
status_summary = pd.DataFrame({
    "Flight Status": status_counts.index,
    "Count": status_counts.values,
    "Percent": status_percentage.values
})

# Plotting the pie chart
fig, ax = plt.subplots()
ax.pie(status_summary['Percent'], labels=status_summary['Flight Status'], autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.show()
```

Highest Flight departures by Country 

```python
import pandas as pd
from tabulate import tabulate

# Load the dataset
data = pd.read_csv('Airline.csv')

# Aggregate the data to count departures per country
country_departures = data['Country Name'].value_counts()

# Calculate the total number of departures
total_departures = country_departures.sum()

# Calculate the percentage of departures per country
country_departures_percent = (country_departures / total_departures) * 100

# Create a DataFrame for display
country_summary = pd.DataFrame({
    "Country Name": country_departures.index,
    "Count": country_departures.values,
    "Percent": country_departures_percent.values
})

# Sort and prepare the DataFrame for the top 10 countries
top_countries_summary = country_summary.sort_values(by="Count", ascending=False).head(10)

# Display the DataFrame in a markdown format using tabulate
print(tabulate(top_countries_summary, headers='keys', tablefmt='pipe', showindex=False))
```

Age Distribution of passengers 

```python
import pandas as pd
from tabulate import tabulate

# Load the dataset
data = pd.read_csv('Airline.csv')

# Ensure the 'Age' column contains numeric data
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')

# Drop any rows where age is NaN (if any)
data = data.dropna(subset=['Age'])

# Define age bins
bins = [0, 18, 30, 40, 50, 60, 70, 80, 100]

# Labels for the age groups
labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-100']

# Use cut to categorize the ages
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# Calculate the number of passengers in each group
age_group_counts = data['Age Group'].value_counts().sort_index()

# Create a DataFrame for display
age_groups_df = pd.DataFrame({
    'Age Group': age_group_counts.index,
    'Number of Passengers': age_group_counts.values
})

# Display the DataFrame using tabulate in a Markdown-friendly format
print(tabulate(age_groups_df, headers='keys', tablefmt='pipe', showindex=False))
```

Visulation of age distribution

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Airline.csv')

# Ensure the 'Age' column contains numeric data
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')

# Drop any rows where age is NaN (if any)
data = data.dropna(subset=['Age'])

# Define age bins
bins = [0, 18, 30, 40, 50, 60, 70, 80, 100]

# Labels for the age groups
labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-100']

# Use cut to categorize the ages
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# Calculate the number of passengers in each group
age_group_counts = data['Age Group'].value_counts().sort_index()

# Create a bar chart for the age groups
plt.figure(figsize=(10, 6))
age_group_counts.plot(kind='bar', color='brown')
plt.title('Distribution of Passengers by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)  # Adjust rotation if needed
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add horizontal grid lines for readability

# Show the plot
plt.show()
```
